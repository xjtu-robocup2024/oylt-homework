```python
import random
import time
import os

class Universe:
    def __init__(self, width=80, height=15):
        self.width = width
        self.height = height
        self.grid = [[' ' for _ in range(width)] for _ in range(height)]
        self.next_grid = [[' ' for _ in range(width)] for _ in range(height)]
    
    def show(self):
        os.system('cls' if os.name == 'nt' else 'clear')
        for row in self.grid:
            print("".join(row))
    
    def Seed(self):
        for i in range(self.height):
            for j in range(self.width):
                if random.random() < 0.25:  # 约25%细胞激活
                    self.grid[i][j] = '*'
    
    def alive(self, x, y):
        if x < 0 or x >= self.height or y < 0 or y >= self.width:
            return False
        return self.grid[x][y] == '*'
    
    def neighbors(self, x, y):
        count = 0
        for i in range(-1, 2):
            for j in range(-1, 2):
                if not (i == 0 and j == 0) and self.alive(x + i, y + j):
                    count += 1
        return count
    
    def next_state(self, x, y):
        live_neighbors = self.neighbors(x, y)
        if self.alive(x, y):
            return '*' if live_neighbors == 2 or live_neighbors == 3 else ' '
        else:
            return '*' if live_neighbors == 3 else ' '
    
    def step(self):
        for i in range(self.height):
            for j in range(self.width):
                self.next_grid[i][j] = self.next_state(i, j)
        self.grid, self.next_grid = self.next_grid, self.grid

def run_game(steps=10, delay=0.2):
    universe = Universe()
    universe.Seed()
    for _ in range(steps):
        universe.show()
        universe.step()
        time.sleep(delay)

if __name__ == "__main__":
    run_game(steps=50, delay=0.2)
```

```python
import random
import time
import os

class Universe:
    def __init__(self, width=80, height=15):
        self.width = width
        self.height = height
        self.grid = [' ' for _ in range(width * height)]
    
    def show(self, generation):
        os.system('cls' if os.name == 'nt' else 'clear')
        print(f"Generation: {generation}")
        for i in range(self.height):
            print("".join(self.grid[i * self.width:(i + 1) * self.width]))
    
    def seed(self):
        for i in range(len(self.grid)):
            if random.random() < 0.25:
                self.grid[i] = '*'
    
    def alive(self, x, y):
        if 0 <= x < self.height and 0 <= y < self.width:
            return self.grid[x * self.width + y] == '*'
        return False
    
    def neighbors(self, x, y):
        count = 0
        for i in range(-1, 2):
            for j in range(-1, 2):
                if (i != 0 or j != 0) and self.alive(x + i, y + j):
                    count += 1
        return count
    
    def step(self):
        changes = []
        for i in range(self.height):
            for j in range(self.width):
                idx = i * self.width + j
                live_neighbors = self.neighbors(i, j)
                if self.grid[idx] == '*' and (live_neighbors < 2 or live_neighbors > 3):
                    changes.append((idx, ' '))
                elif self.grid[idx] == ' ' and live_neighbors == 3:
                    changes.append((idx, '*'))
        
        # 更新细胞状态
        for idx, state in changes:
            self.grid[idx] = state

def run_game(steps=10, delay=5.0):
    universe = Universe()
    universe.seed()
    for gen in range(steps):
        universe.show(gen)
        universe.step()
        time.sleep(delay)

if __name__ == "__main__":
    run_game(steps=50, delay=5.0)
```
